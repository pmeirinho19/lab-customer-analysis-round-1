```python
#5

df1 = pd.concat([file1, file2, file3], axis=0)

df1.head(10)
```

```python

#6


# Numerical Columns: "Costumer Lifetime Value" , "Income", "Monthly Premium Auto", "Total Claim Amount", Number of Open Complains

# Categorial Columns: "Customer", "State", "Gender", "Education", "Policy Type", "Vehicle Class",
```

```python
#7

df1 = df1.drop(labels='Education', axis=1)
df1 = df1.drop(labels='Number of Open Complaints', axis=1)

df1
```

```python
# It is need to turn the values of "Customuer life Value into a number" but first:
#Is need to delete the symbol%
```

```python
df1['Customer Lifetime Value']
```

```python
df1 = df1.reset_index(drop=True)
```

```python
def clean_perc_symbol(x):
    if type(x) == str:
        x = float(x.replace("%"," "))
    return x
```

```python
df1['Customer Lifetime Value'] = df1['Customer Lifetime Value'].apply(clean_perc_symbol)
```

```python
def dividing_for_100(x):
    y = x/100
    return y
```

```python
df1['Customer Lifetime Value'] = df1['Customer Lifetime Value'].apply(dividing_for_100)
```
